```python
# Merging products dataframe and aisles dataframe based on common column aisle_id to create temp dataframe merge1
merge1 = pd.merge(prod, aisles)
```

```python
# Merging merge1 dataframe and dept dataframe based on common column department_id to create temp dataframe merge2
merge2 = pd.merge(merge1, dept)
```

```python
# merge temp dataframe merge2 and order_prod dataframe on common column product_id to create full dataframe insta
insta = pd.merge(merge2, order_prod)
```

```python
# Which orders had the most items ordered
insta.groupby('order_id').order_id.count().sort_values(ascending = False).head()
```

```python
# Top items that were added to cart first by user
insta[insta.add_to_cart_order == 1].groupby('product_name').product_name.count().sort_values(ascending = False).head(100)
```

```python
top_aisle = print(insta.groupby('aisle').product_name.head(1))
```

```python
insta.loc[insta['product_name'].isin(top_aisle)].sort_values('aisle')
```

```python
print(insta.groupby('aisle').aisle.count().sort_values(ascending = False).head(20))
insta.groupby('aisle').aisle.count().sort_values(ascending = False).head(20).plot(kind='bar');
plt.title('Top 20 Aisles by Order')
```

```python
print(insta.groupby('department').department.count().sort_values(ascending = False).head(20))
insta.groupby('department').department.count().sort_values(ascending = False).head(20).plot(kind='bar');
plt.title('Top 20 Departments by Order')
```
